```python
# set operations

# union - combine sets
rainbow = {'red', 'orange', 'yellow', 'green','blue', 'indigo', 'violet'}
olympic_flag = {'red', 'green', 'yellow', 'blue','black'}
rainbow.union(olympic_flag)
#{'black', 'blue', 'green', 'indigo', 'orange','red', 'violet', 'yellow'}

# intersection - finds common values only
rainbow = {'red', 'orange', 'yellow', 'green','blue', 'indigo', 'violet'}
olympic_flag = {'red', 'green', 'yellow', 'blue','black'}
rainbow.intersection(olympic_flag) #{'blue', 'green', 'red', 'yellow'}

# symmetric difference -> items in one but not both sets
rainbow = {'red', 'orange', 'yellow', 'green','blue', 'indigo', 'violet'}
olympic_flag = {'red', 'green', 'yellow', 'blue','black'}
rainbow.symmetric_difference(olympic_flag) #{‘black’, ‘indigo’, ‘orange’, ‘violet'}

# difference - items in first set, but not second
rainbow = {'red', 'orange', 'yellow', 'green','blue', 'indigo', 'violet'}
olympic_flag = {'red', 'green', 'yellow', 'blue', 'black'}
rainbow.difference(olympic_flag) #{‘orange’, ‘indigo’, ‘violet’}
olympic_flag.difference(rainbow) #{‘black’}
```

# Dictionaries

```python
# Example of an empty dictionary
empty_dictionary = {}

# Examples of a dictionary
capitals = {'Canada': 'Ottawa',
    'United States': 'Washington, D.C.',
    'Mexico': 'Mexico City'}
```

```python
# key access
capitals['Canada'] #returns ‘Ottawa’

# key error example
# capitals['Italy'] #returns KeyError: 'Italy'
```

```python
# accessing dictionaries

# Example
capitals.get('Italy', 'Not Found') # returns the value, or 'Not Found' as a placeholder
#returns ‘Not Found’

# Example
'United States' in capitals # True

'Washington, D.C.' in capitals # False; in looks in keys, not values
```

```python
# updating dictionaries

# assigning to key not in dict adds the key:value pair
capitals['Italy'] = 'Milan'
capitals # {'Canada': 'Ottawa', 'United States':'Washington, D.C.', 'Mexico': 'Mexico City','Italy': ‘Milan'}

# assigning to existing key overwrites value
capitals['Italy'] = 'Rome'
capitals # {'Canada': 'Ottawa', 'United States':'Washington, D.C.', 'Mexico': 'Mexico City','Italy’: ‘Rome'}
```

```python
# removing key:value

del capitals['Mexico']
capitals # {'Canada': 'Ottawa', 'United States':'Washington, D.C.', 'Italy': ‘Milan'}
```

```python
# dictionary methods

capitals.keys() # dict_keys(['Canada', 'United States', 'Italy'])

if 'Ottawa' in capitals.values():
    print('Ottawa is a capital')
else:
    print('Ottawa is not a capital')
```

# Iterations and loops

```python
# for loops
vowels =['a', 'e', 'i', 'o', 'u']
for vowel in vowels:
    print(vowel) # uoiea
```

```python
# range

for i in range(5):
    print(i, i*2)
# 0 0
# 1 2
# 2 4
# 3 6
# 4 8
```

```python
# loops and lists

input_files = ['data_01.csv', 'data_02.csv', 'data_03.csv','data_04.csv']
output_files = []
for i in input_files:
    output_file_name = 'processed_' + i.replace('.csv', '.xlsx')
    output_files.append(output_file_name)

output_files # ['processed_data_01.xlsx','processed_data_02.xlsx','processed_data_03.xlsx','processed_data_04.xlsx']
```

```python
# looping with multiple values

stops = ['Sheppard-Yonge', 'Bayview', 'Bessarion', 'Leslie','Don Mills']

# enumerate()
# gives two values for each item: the index and the value itself.

for idx, stop in enumerate(stops):
    print(f'Stop {idx + 1} is {stop}.')
# Stop 1 is Sheppard-Yonge.
# Stop 2 is Bayview.
# Stop 3 is Bessarion.
# Stop 4 is Leslie.
# Stop 5 is Don Mills.
```

```python
# dict.items() 
# items() gives two values for each item: the key and the value itself.

prices = {"apple": 2, "banana": 1, "orange": 3}
for fruit, price in prices.items():
    print(fruit, price)
# apple 2
# banana 1
# orange 3
```

```python
# nested loops
for i in range(1, 3): # outer loop = 1
    for j in range(1, 3): # inner loop = 1
        print(i, "*", j, "=", i*j)
# 1 * 1 = 1
```

```python
for i in range(1,3):
    for j in range(1,10):
        print(f'outer {i}, inner {j}')
```

# while loops

```python
countdown = 4
while countdown > 0:
    print (countdown)
    countdown -= 1
```

```python
# break

countdown = 4
while countdown > 0:
    print (countdown)
    if countdown == 3:
        break
    countdown -= 1
```

```python
# continue

for i in range(1, 6):
    if i % 2 == 0: # if the number is even
        continue # skip the rest of the loop for this iteration
    print(i)
```
