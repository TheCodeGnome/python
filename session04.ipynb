# Reading and writing files 

# Navigating folders

```python
import os
```

```python
# get path to the folder we’re currently in
os.getcwd()
# see contents of the current folder
os.listdir()
# change to a new directory
os.chdir()
```

```python
# Path to the folder you want to look inside
folder_path = './05_src/data/slides_data'
# Get a list of all files and folders inside
items = os.listdir(folder_path)
# Print each item
for item in items:
    print(item)
```

```python
# open(file_path, mode) -> mode = (r)ead, (w)rite, (a)ppend
file_path = './05_src/data/slides_data/california_housing_test.csv'
with open(file_path, 'r') as file:
    print(file) # gives header info

with open(file_path, 'r') as file:
    # print(file.read()) # output whole file
    print(file.readline()) # returns only the next line. Python tracks index.
    print(file.readline()) # returns only the next line. Python tracks index.
    # print(file.readlines()) # outputs each line as an item in a list
```

# Writing to file

```python
# write() - writes single string to file
# writelines() - writes a list of strings to a file; need to separate new lines

airlines = [
    "Code\tAirline\n",
    "AAL\tAmerican Airlines\n",
    "ABK\tAirbus Canada\n",
    "ABP\tABS Jets\n",
    "ACA\tAir Canada\n"
]
with open('airlines.txt', 'w') as file: # 'w' will overwrite
    file.writelines(airlines)

with open('airlines.txt', 'a') as file:
    file.write("ARQ\tAirico\n")
```
