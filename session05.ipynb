## Numpy

```python
import numpy as np
```

```python
# Python list
numbers_list = [1, 2, 3, 4, 5]
# NumPy array
numbers_array = np.array([1, 2, 3, 4, 5])
```

```python
# 1 row x 5 columns - vector, 1d
vector = np.array([1, 2, 3, 4, 5])
# vector.ndim = 1 -> axes
# vector.shape = (5,1) -> size in each direction
# vector.size = 5 -> number of values

# 2 rows x 3 columns; matrix - 2d
matrix = np.array([[1, 2, 3],
[4, 5, 6]])
matrix
# matrix.ndim = 2, matrix.shape = (2,3), matrix.size = 6
```

```python
# Placeholder arrays
np.zeros((2,3)) # all zeros

np.ones((2,3)) # all 1s
```

```python
# special numpy arrays

# NumPy array from 1 until 10 in steps of 2
np.arange(1,10,2)

# NumPy array of 5 random integers between 1 and 10
np.random.seed(1234) # makes the random set repeatable 
random_array = np.random.randint(1,10,5)
random_array2 = np.random.randint(1,10,5)
print(random_array)
print(random_array2)

# NumPy array of 10 repeated 5 times
np.repeat(10,5)

# NumPy array repeated x times
np.tile(vector, 2)
```

```python
# simulating data

# Create a Generator
rng = np.random.default_rng()

# 10 data points from normal distributed data with a mean of 0, sd 1
rng.normal(0,1,10)
# can also use uniform distribution, poisson etc
```

```python
# reshaping arrays

# 2 rows x 3 columns
matrix = np.array([[1, 2, 3],[4, 5, 6]])

# 3 rows x 2 column
matrix.T # transpose
```

```python
# collapse/flatten to 1d

# 2 rows x 3 columns
matrix = np.array([[1, 2, 3],[4, 5, 6]])
# 1 row x 6 columns
matrix.flatten()
```

```python
# shared axes size, can be stacked together on that axes

matrix1 = np.array([[1, 2, 3],
[4, 5, 6]])
matrix2 = np.array([[7, 8, 9],
[10, 11, 12]])

print(np.vstack((matrix1,matrix2))) #vertical stack
print(np.hstack((matrix1,matrix2))) #horizontal stack

```

```python
# apply operations to all values at once, without a loop

numbers_array = np.array([1, 2, 3, 4, 5])
numbers_array * 2
print(numbers_array)
# array([ 2, 4, 6, 8, 10])

# can perform operations when arrays are same length
matrix1 = np.array([[1, 2, 3],[4, 5, 6]])
matrix2 = np.array([[7, 8, 9],[10, 11, 12]])
matrix_sum = matrix2 - matrix1
print(matrix_sum)

# can summarize values
matrix1 = np.array([[1, 2, 3],[4, 5, 6]])
print(matrix1.sum()) #1 + 2 + 3 + 4 + 5 + 6 = 21
print(matrix1.max()) #6
print(matrix1.mean()) # 1 + 2 + 3 + 4 + 5 + 6 / 6 = 3.5
```

```python
# mutate or update elements; also, copy
matrix1 = np.array([[1, 2, 3],[4, 5, 6]])
matrix3 = matrix1.copy()
matrix3[1,1] = 50 #replace 5 with 50, original not changed
print(matrix3)
print(matrix1)
```

```python
# check if all values share factor
print(matrix3 % 3 == 0)

# can filter with boolean expression as mask
matrix3[matrix3 % 3 == 0]
```

```python
# masks to generate new arrays with conditionals
np.where(matrix3 % 3 == 0, # condition
matrix3, # return the element if True
0) # return 0 if False
# array([[ 0, 0, 3], [ 0, 0, 6]])
```

## Pandas

- Python library for data analysis and manipulation.
- Works with tabular data (like csv)
- Includes dataframes (2d labeled tables), series (1d labelled array)

```python
import pandas as pd
```

## Dataframes

```python
# cols labelled with names
# rows have label or index
# each col is a Series (1d array) with shared type
# can have different types across cols

df = pd.DataFrame({
    "X": [True, False, True, True, False], # bool
    "Y": [3.14, 7.65, 2.56, 7.89, 1.23], # float
    "Z": ["o", "a-", "b+", "b-", "o"] # str
})
```

```python
bike_thefts_df = pd.read_csv('./05_src/data/slides_data/bicycle_thefts_4326.csv')
print(bike_thefts_df.shape)
print(bike_thefts_df.dtypes)
```

```python
print(bike_thefts_df.head()) # defaults to 5 records
```

```python
print(bike_thefts_df.tail(3)) # last 3 items
```

```python
# Renaming columns

bike_thefts_df = bike_thefts_df.rename(columns = {"cost_of_bike": "bike_cost" })
```

```python
# Profiling columns
# focus on subset of data; can get single column by name like a key
df = bike_thefts_df
df["Status"].unique()
# Output
# array(['STOLEN', 'RECOVERED', 'UNKNOWN'], dtype=object)

# count occurrences
df["Status"].value_counts()
```

```python
# get overview of dataframe; shape, size (rows and cols), keys and datatype, count of non-nulls in each col
df.info()
```

## Changing datatypes

```python
# Convert occurrence date columns to datetime
df["Occurrence_Date"] = pd.to_datetime(df["Occurrence_Date"])

# Convert Status columns to category
df["Status"] = df["Status"].astype("category")
```

## Describing and editing dataframe

```python
df.describe(include = "all") # by default, numeric cols only; use include="all" to get all cols
```

```python
# drop cols; won't affect original dataframe unless re-assigned
df.drop(columns = "_id")

# add new col with repeated data
df["current_year"] = 2026

# add col with parts of existing col
df["Occurrence_Month"] = df["Occurrence_Month"].str[:3]
```

## Filter and select

```python
# loc access rows based on bool
df.loc[df["Bike_Model"].isna()]
```

```python
# access rows of data where line is NA; query() is easier to read the code, but is slower than loc()
df.query("Bike_Model.isna()")
```

```python
# Filtering

Models_to_keep = ["HYBRID","SIRRUS","ESCAPE"]
# Filter using .loc[]
df.loc[df["Bike_Model"].isin(Models_to_keep)]
```

```python
# large scale data edit

#This makes all the upper case A’s lowercase
df["Bike_Model"] = (df["Bike_Model"].str.replace("A", "a"))
```

## Grouping

```python
# core workflow in pandas is split-apply-combine; group, apply func to group, combine result into a data structure

#Let’s first group the data by bike model
Model_Groups = df.groupby("Bike_Model")

#Identify unique manufacturers for each group
Model_Groups["Bike_Make"].unique()

# write dataframe to file
df.to_csv("updated_df.csv")
```
